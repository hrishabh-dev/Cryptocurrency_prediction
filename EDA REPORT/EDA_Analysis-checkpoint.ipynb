## there are some outliers in market cap and 24 hours volume we will treat that but generally we don't do that especially in crypto market

```python
## removing outliers
df['mkt_cap']=df['mkt_cap'].clip(upper=df['mkt_cap'].quantile(0.99))
```

```python
## again checking if it worked
```

```python
sns.boxplot(df)
```

```python
## we will use iqr method to remove outliers
```

```python
# Calculate the IQR
Q1 = df['24h_volume'].quantile(0.25)
Q3 = df['24h_volume'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Filter out outliers
df = df[(df['24h_volume'] >= lower_bound) & (df['24h_volume'] <= upper_bound)]

# Visualize the data with a box plot
sns.boxplot(x=df['24h_volume'])
plt.title('Box Plot of 24h_volume (Outliers Removed)')
plt.show()
```

```python
# Calculate the IQR for 'mkt_cap'
Q1_mkt_cap = df['mkt_cap'].quantile(0.25)
Q3_mkt_cap = df['mkt_cap'].quantile(0.75)
IQR_mkt_cap = Q3_mkt_cap - Q1_mkt_cap

# Define upper and lower bounds for outliers in 'mkt_cap'
upper_bound_mkt_cap = Q3_mkt_cap + 1.5 * IQR_mkt_cap
lower_bound_mkt_cap = Q1_mkt_cap - 1.5 * IQR_mkt_cap

# Filter out outliers from 'mkt_cap'
df = df[(df['mkt_cap'] >= lower_bound_mkt_cap) & (df['mkt_cap'] <= upper_bound_mkt_cap)]

# Visualize the data with a box plot for 'mkt_cap'
sns.boxplot(x=df['mkt_cap'])
plt.title('Box Plot of mkt_cap (Outliers Removed)')
plt.show()
```

```python
sns.boxplot(df)
```

##  used the IQR method and clipping to address outliers in 'mkt_cap' and '24h_volume'. This is generally acceptable for taming extreme values, but it's crucial to acknowledge that removing these data points might discard valuable information about significant market events

## now i will use function to remove outliers because individually it wil take time

```python
# Function to remove outliers using IQR method
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.20)
    Q3 = data[column].quantile(0.90)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data_filtered = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data_filtered


df_cleaned = remove_outliers_iqr(df, 'mkt_cap')
df_cleaned = remove_outliers_iqr(df_cleaned, '24h_volume')
df_cleaned = remove_outliers_iqr(df_cleaned, '1h')
df_cleaned = remove_outliers_iqr(df_cleaned, '24h')
df_cleaned = remove_outliers_iqr(df_cleaned, '7d')
df_cleaned = remove_outliers_iqr(df_cleaned, 'price')

```

## Despite applying outlier removal techniques like the IQR method and clipping, some outliers may persist in the dataset. This is because financial market data, especially in the cryptocurrency domain, is inherently volatile and prone to extreme fluctuations. These fluctuations can be driven by various factors, including market sentiment, regulatory news, and technological advancements, leading to legitimate but extreme data points that are difficult to completely eliminate without potentially discarding valuable information about significant market events

#### 1. Distribution of Individual Variables:

The diagonal plots of the pair plot show the Kernel Density Estimation (KDE) for each individual variable (e.g., 'price', 'mkt_cap', '24h_volume', etc.).
The shape of the KDE curve reveals the distribution of the variable. For example, a bell-shaped curve suggests a normal distribution, while a skewed curve indicates a non-normal distribution. You can observe the skewness and kurtosis of each variable's distribution.

#### 2. Relationships between Variables:

The off-diagonal plots illustrate the relationship between pairs of variables. The density of points in a particular region of the plot indicates the likelihood of observing that combination of values.
For example, if the density is concentrated along a diagonal line, it suggests a positive correlation between the two variables. If the density is scattered, it indicates a weak or no correlation

The diagonal plots reveal the individual distributions of variables such as price, market capitalization, and 24-hour volume, highlighting their skewness and kurtosis. Off-diagonal plots illustrate the relationships between pairs of variables, where high-density regions indicate a greater likelihood of observing specific value combinations. These plots help identify potential correlations, with density concentrated along a diagonal suggesting a positive relationship, and scattered density indicating a weak or no correlation. By analyzing these density plots, we gain a comprehensive understanding of the data's characteristics, including variable distributions, interdependencies, and potential outliers or clusters

# Time Series Analysis

#Since my dataset likely contains temporal information (dates/timestamps), exploring time series analysis can provide additional insights into the behavior of cryptocurrency prices and volumes over time.

```python
# 1. Convert the 'date' column to datetime objects if it's not already.
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'])

# 2. Set the 'date' column as the index for time series operations.
df_cleaned.set_index('date', inplace=True)

# 3. Plot the 'price' over time.
plt.figure(figsize=(12, 6))
plt.plot(df_cleaned['price'])
plt.title('Cryptocurrency Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()

# 4. Analyze rolling statistics (e.g., moving averages).
df_cleaned['price_rolling_mean'] = df_cleaned['price'].rolling(window=7).mean()  # 7-day rolling mean
plt.figure(figsize=(12, 6))
plt.plot(df_cleaned['price'], label='Price')
plt.plot(df_cleaned['price_rolling_mean'], label='7-Day Rolling Mean')
plt.title('Cryptocurrency Price with Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

## The time series analysis reveals that the cryptocurrency price exhibits significant fluctuations over the observed period. The initial plot of the price highlights the volatility inherent in the cryptocurrency market. To discern underlying trends, a 7-day rolling mean was calculated and plotted alongside the raw price data. This rolling average smooths out short-term price movements, revealing a clearer picture of the overall price trend

## Disregarding retail investors, this analysis highlights the dominant role of large market participants ("whales"). They strategically accumulate cryptocurrencies at lower prices, resulting in substantial market capitalization despite modest price fluctuations within 7 days and 24 hours. This pattern, evident in scatter plots where low prices coincide with high market capitalization, underscores their calculated approach. While correlation between 7-day price changes and market capitalization is moderate, it signifies big players' influence on medium-term trends. Ultimately, these influential entities significantly shape cryptocurrency market dynamics.
