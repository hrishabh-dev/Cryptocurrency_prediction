### Feature engineering is the process of creating new features from existing data to improve the performance of machine learning models

###  In this case, liquidity was calculated as a new feature because it is a key indicator of market dynamics and can significantly impact cryptocurrency prices and trading behavior. By adding liquidity as a feature, we provide the model with valuable information about the ease with which a coin can be bought or sold, which can help predict future price movements and inform trading decisions

```python
df['liquidity'] = df['24h_volume'] / df['mkt_cap']
```
